```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(gapminder[['fertility']], gapminder['life'], test_size=0.2, random_state=42)

# Train a linear regression model with the 'fertility' feature
lr = LinearRegression()
lr.fit(X_train, y_train)

# Predict life expectancy for the test data
y_pred = lr.predict(X_test)

# Evaluate the model using R-squared score
score = lr.score(X_test, y_test)
print(f"Model score with 'fertility' feature: {score:.3f}")

```
